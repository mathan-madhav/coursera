## Introduction

ABC Petrol Company wants to launch a Petrol pump in some locality in Chennai, a populous metro-politician city in India. The company approaches me to suggest the best location for the new petrol pump so that the profitability can be increased.

## Problem Background:


To suggest the best locality for ABC Petrol company, we need to gather the localities of Chennai city and analyze the places where target customers are high in numbers so that the company can gain more profit.


## Problem Description:


In order to find the suitable place to launch a petrol pump we need to analyze the neighborhood localities in Chennai city and apply the following logics.

1. The primary logic is to identify the roads connecting highways and localities with higher number of offices and buildings.

2. The secondary logic is to find the locations of other petrol pumps and the locality with sparse number of petrol pumps will have higher opportunity to grab the attention of petrol consumers.



## Data Section:

ABC Petrol Company wants to launch a Petrol pump in some locality in Chennai city. The company approaches me to suggest the best location for the new petrol pump so that the profitability can be increased.

So, for this problem, we must collect the geo-coordinates of the neighborhood localities in the city of Chennai and as well as the geo-coordinates of Offices, other petrol pumps and the roads (meeting junctions) where the motion of vehicles is quite more.

## Fetch Data

```python
## Import Section

import pandas as pd
import io
import requests
import matplotlib.pyplot as plt
```

```python
# Fetch Localities data

url = 'https://github.com/mathan-madhav/coursera/blob/master/data_localities_coords.csv'
s=requests.get(url).content
localities=dict(pd.read_csv(io.StringIO(s.decode('utf-8'))))
print(localities)
```

```python
# Visualize Localities

locality_x = []
locality_y = []

# Get Latitude and Longitude

for loc in localities.keys():
    locality_x.append(localities[loc][0])
    locality_y.append(localities[loc][1])

# Round up the data for effective visualization

locality_x = [round(float(i),1) for i in locality_x[1:len(locality_x)]]
locality_y = [round(float(i),1) for i in locality_y[1:len(locality_y)]]

# Draw maps

ax=plt.gca()
plt.xlabel('Latitude')
plt.ylabel('Longitude')

plt.plot(locality_x,locality_y,'ro',label='Localities')
plt.legend(loc='best')
plt.title('Localities in Chennai City')
```

```python
# Fetch Chief centers

url = 'https://github.com/mathan-madhav/coursera/blob/master/data_chief_centers.csv'
s=requests.get(url).content
centers=dict(pd.read_csv(io.StringIO(s.decode('utf-8'))))
print(centers)
```
